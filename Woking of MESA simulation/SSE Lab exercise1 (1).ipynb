```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'profile3.data'

with open(file_path, 'r') as f:
    lines = f.readlines()

for i, line in enumerate(lines):
    if 'logR' in line:
        header_line_index = i
        headers = line.strip().split()
        break

df = pd.read_csv(
    file_path,
    skiprows=header_line_index + 1,
    delim_whitespace=True,
    names=headers
)

radius = 10 ** df['logR'].to_numpy()  
delta_radius = np.diff(radius)     

plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, len(delta_radius) + 1), delta_radius, lw=2.5) 

plt.gca().set_facecolor('white')

plt.xlabel('Zone Number', fontsize=16) 
plt.ylabel('Radius', fontsize=16)
plt.title('Grid Spacing in Stellar Profile', fontsize=18)

plt.tick_params(axis='both', which='major', labelsize=14)

plt.grid(False)
plt.tight_layout()
plt.savefig("ex_1_q3.jpg", dpi = 1200)
plt.show()

if radius[0] < radius[-1]:
    print("Zone 1 is the central zone.")
else:
    print("Last zone is the central zone.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

history_path = "history.data"

with open(history_path, 'r') as f:
    lines = f.readlines()

for i, line in enumerate(lines):
    if 'star_age' in line and not line.startswith('#'):
        header_index = i
        column_names = line.strip().split()
        break

df = pd.read_csv(history_path, skiprows=header_index + 1, delim_whitespace=True, names=column_names)

model_number = df['model_number'].to_numpy()
star_age = df['star_age'].to_numpy()
radius = 10 ** df['log_R'].to_numpy()  
luminosity = 10 ** df['log_L'].to_numpy()  
mass = df['star_mass'].to_numpy() 

# cons
R_sun = 6.957e10        # cm
M_sun = 1.989e33        # g
L_sun = 3.828e33        # erg/s
G = 6.6743e-8           # cm^3 g^-1 s^-2

R = radius * R_sun
M = mass * M_sun
L = luminosity * L_sun

dR_dt = np.gradient(R, star_age * 3.154e7)
tau_contraction = np.abs(R / dR_dt)

tau_dyn = np.sqrt(R**3 / (G * M))

tau_thermal = ((G * M**2) / (R * L))*(1/2)

c = 3.0e10  # speed cm/s
tau_nuclear = (0.01 * M * c**2) / L

sec_to_years = 1 / 3.154e7
tau_contraction /= 3.154e7
tau_dyn /= 3.154e7
tau_thermal /= 3.154e7
tau_nuclear /= 3.154e7

plt.figure(figsize=(12, 7))
plt.plot(model_number, tau_contraction, label=r"$\frac{R}{{R} \odot}$ (Expansion/Contraction)", lw=2.5)
plt.plot(model_number, tau_dyn, label='Dynamical Timescale', lw=2.5)
plt.plot(model_number, tau_thermal, label='Thermal Timescale', lw=2.5)
plt.plot(model_number, tau_nuclear, label='Nuclear Timescale', lw=2.5)

plt.yscale('log')

plt.xlabel('Model Number', fontsize=16)
plt.ylabel('Timescale (years)', fontsize=16)
plt.title('Stellar Timescales During Evolution', fontsize=18)

plt.tick_params(axis='both', which='major', labelsize=14)

plt.legend(fontsize = 16)
plt.grid(False)
plt.tight_layout()
#plt.savefig("ex_1_q4.jpg", dpi = 1200)
plt.text(200,1, rf"$T_{{\rm Dyn}} = \frac{{\rm 0.02}}{{\rm 360}} ( \frac{{\rm R}}{{\rm R \odot}})^{{\rm 3/2}}$", fontsize = 22)
plt.text(200,100000000, 
         rf"$T_{{\rm Nuc}} = 10^{{\rm 10}} \frac{{\rm M}}{{\rm M \odot}} ( \frac{{\rm R}}{{\rm R \odot}})^{{\rm 3/2}}$", fontsize = 22)
plt.text(300,10000, 
         rf"$T_{{\rm Ther}} = 1.5 \times 10^{{\rm 7}} \frac{{\rm R  L\odot}}{{\rm R\odot  L}} $", fontsize = 22)

plt.savefig("ex_1_q4.jpg", dpi = 800)
plt.show()

display(Math(r"\tau_{contraction} = \frac{|R|}{|\dot{R}|}"))
display(Math(r"\tau_{dyn} = \sqrt{\frac{R^3}{G \cdot M}}"))
display(Math(r"\tau_{thermal} = \frac{G \cdot M^2}{R \cdot L}"))
display(Math(r"\tau_{nuclear} = \frac{0.1 \cdot M \cdot c^2}{L}"))
```
